```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# establish a folder named LAB1
path = "LAB1"
if not os.path.isdir(path):
    os.mkdir(path)
```

```python
# RGB picture showing
imgCV = cv2.imread('dumplings.jpg')              # read picture and change it to ndarray
imgRGB = cv2.cvtColor(imgCV, cv2.COLOR_BGR2RGB)  # openCV is BGR, plt.imshow is RGB. Need change BGR to RGB
plt.figure(figsize=(30,10)) 
plt.imshow(imgRGB)
# save picture
cv2.imwrite('LAB1/dumplingsRGB.jpg', imgCV)
```

```python
# Gray picture showing
imgGray = cv2.cvtColor(imgRGB,cv2.COLOR_BGR2GRAY)   # change BGR to gray
plt.figure(figsize=(30,10)) 
plt.imshow(imgGray, cmap = 'gray')  # remember to add cmap = 'gray'
# save picture
cv2.imwrite('LAB1/dumplingsGray.jpg', imgGray)
```

```python
# find the biggest pixel
print(imgGray)
print('Max Pixel Value :',imgGray.max())
```

```python
# Anti White
maxPixel = imgGray.max()
imgAntiWhite = maxPixel-imgGray
print(imgAntiWhite)
```

```python
# Anti White gray picture showing
plt.figure(figsize=(30,10)) 
plt.imshow(imgAntiWhite, cmap = 'gray') # remember to add cmap = 'gray'
# save picture
cv2.imwrite('LAB1/dumplingsAntiWhite.jpg', imgAntiWhite)
```

```python
# find the mean pixel
meanPixel = imgGray.mean()
print('Mean Pixel Value :',meanPixel)
```

```python
# Binarization picture showing
imgBinarization =  imgGray > meanPixel
print(imgBinarization)
plt.figure(figsize=(30,10)) 
plt.imshow(imgBinarization, cmap = 'gray')  # remember to add cmap = 'gray'
# save picture
cv2.imwrite('LAB1/dumplingsBinarization.jpg', (imgBinarization + 0)*255)
```
